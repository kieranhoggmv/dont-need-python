```python
import pandas as pd

df = pd.read_csv("combined.csv", index_col=0)
df.head()
```

```python
# Fix Student IDs
df["Student ID"] = df["Student ID"].astype(str)
df["Student ID"] = df["Student ID"].apply(lambda x: "0" + x if len(x) == 11 else x)
df[df["Student ID"].apply(lambda x: len(x) == 11)]
```

```python
df["Test Date"] = pd.to_datetime(df["Test Date"], format="mixed")
df["Test Date"] = df["Test Date"].dt.date
```

```python
# df = df.set_index("Student ID")
summary_df = df.pivot_table(values=["Standardised Score"], index=["Student ID"], columns="Test Name")
summary_df.columns = summary_df.columns.to_flat_index()
df.set_index("Student ID")[["Test Date", "Gender", "Name", "Email"]].join(summary_df)
```

```python
df.to_csv("done.csv", index=False)
```
